### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

Syntax:
`collection.find(filter, projection)`

```python
# Your Code
filter = {"number_of_employees": {"$gt": 500}}
projection = {"_id":0, "name":1, "number_of_employees":1 }
```

```python
list(c.find(filter, projection).sort("number_of_employees", -1).limit(20))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
# Your Code
filter = {"founded_year": {"$gte": 2000, "$lte":2005}}
projection = {"_id":0, "name":1, "founded_year":1 }
list(c.find(filter, projection).sort("founded_year", 1))
```

```python
len(list(c.find(filter, projection).sort("founded_year", 1)))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
# Your Code
filter = {"$and": 
          [{"ipo.valuation_amount": {"$gt": 100000000}},   
        {"founded_year": {"$lt": 2010}}, 
        {"ipo.valuation_amount": {"$ne": None}},   
        {"ipo.valuation_amount": {"$exists": True}}]}# valuation amount no tiene que ser None y tiene que existir
projection = {"_id":0, "name":1, "ipo.valuation_amount":1 }
list(c.find(filter, projection).sort("ipo.valuation_amount", -1))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
# Your Code
employees = {"number_of_employees": {"$lt": 1000}}
founded = {"founded_year": {"$lt":2005}}
projection = {"_id":0, "name":1, "number_of_employees":1, "founded_year":1 }
list(c.find({"$and":[employees, founded]},projection).sort("number_of_employees",-1).limit(10))
```

### BONUSASSO 20. All the companies that have been 'deadpooled' after the third year. 
Tip: You should use a new operator. Try to search WHERE the solution is. 

```python
# Your Code
# where se usa con código Java Script por eso hay que añadir el this.
condition1= {"$where": "this.deadpooled_year - this.founded_year > 3"} # miro que el deadpolled year sea 4 años después de la fundación
condition2= {"$and":[
        {"founded_year": {"$ne": None, "$exists": True}}]}# filtro para que founded_year no sea nulo y exista
projection= {"_id":0, "name":1, "deadpooled_year":1, "founded_year":1}
list(c.find({"$and":[condition1, condition2]},projection))
```
